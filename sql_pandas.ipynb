**Automobile dataset...**

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Automobile_data.csv")
df
```

```python
df1 = df[['company','price']][df.price==df['price'].max()]
df1
```

```python

# Group the original DataFrame by 'company'
car = df.groupby('company')

# Get the group corresponding to 'toyota' (note the spelling)
df_toyota = car.get_group('toyota')
df_toyota
```

```python
totalcar=car.size()
totalcar
```

```python
carprice=car['price'].max()
carprice
```

```python
avgmileg=car['average-mileage'].mean()
avgmileg
```

**occupation dataset...**

```python
import pandas as pd
occ = pd.read_csv("/content/drive/MyDrive/occupation.csv")
occ
```

```python
avgocc=occ.groupby('occupation')['age'].mean()
avgocc
```

```python
maxocc,minocc=occ.groupby('occupation')['age'].max(),occ.groupby('occupation')['age'].min()
print(maxocc)
print("\n Min ages\n")
print(minocc)
```

```python
out5 = occ.groupby(by=["occupation","gender"])[["age"]].mean().add_prefix("MeanOf_").reset_index()
out6 = out5.pivot(index="occupation",columns="gender",values="MeanOf_age")
out6
```

```python
ratioocc=ratioocc = occ[occ['gender'] == 'M'].groupby('occupation')['gender'].count()
ratioocc
```
